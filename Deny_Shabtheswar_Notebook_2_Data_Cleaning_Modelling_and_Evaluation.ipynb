Code adapted from Code Reuse Session 2 – Load Cleaned Dataset

```python
# Import pandas for data handling
import pandas as pd

# Load the cleaned classification CSV produced in Notebook 1
df_class = pd.read_csv('/content/drive/MyDrive/df_class_clean.csv')

# Display the first five rows to verify the dataset
df_class.head()
```

Code adapted from Code Reuse Session 2 – Dataset Overview

```python
# Show number of rows and columns
df_class.shape
```

```python
# Show list of columns
df_class.columns
```

Code adapted from Code Reuse Session 2 – Define X and y

```python
# Define target variable y
y = df_class['loan_approval_status']

# Define feature set X by dropping the target column
X = df_class.drop(columns=['loan_approval_status'])
```

```python
# 1. Print the list of feature names used for modelling
print("List of Feature Names Used (X):")
print(X.columns.tolist())
```

```python
# 2. Print the data shape output
print("\nData Shape Function Output:")
print("Features (X) Shape:", X.shape)
print("Target (y) Shape:", y.shape)
```

Code adapted from Code Reuse Session 2 – Train/Test Split (stratified)

```python
# Import train_test_split
from sklearn.model_selection import train_test_split

# Perform an 80/20 split with stratification and fixed random state
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

# Proving the Accept/Reject ratio is the same
print("Training Set Ratio:\n", y_train.value_counts(normalize=True))
print("\nTest Set Ratio:\n", y_test.value_counts(normalize=True))
```

Code adapted from Code Reuse Session 2 – Identify Column Types

```python
# Show data types
X_train.dtypes

# Create a list of numeric columns (float / int)
numeric_cols = X_train.select_dtypes(include=['int64','float64']).columns.tolist()

# Create a list of categorical columns (object / category)
categorical_cols = X_train.select_dtypes(include=['object','category']).columns.tolist()

# Print the two lists for confirmation
numeric_cols, categorical_cols
```

Code adapted from Code Reuse Session 2 – Manual Categorical Encoding (get_dummies)

```python
# Create one-hot encoding of categorical variables on training set
X_train_cat = pd.get_dummies(X_train[categorical_cols], drop_first=True)

# Create one-hot encoding of categorical variables on test set
X_test_cat = pd.get_dummies(X_test[categorical_cols], drop_first=True)

# Align test columns to training columns (add missing columns with zeros)
X_train_cat, X_test_cat = X_train_cat.align(X_test_cat, join='outer', axis=1, fill_value=0)
```

```python
#Display X_train_cat
X_train_cat.head()
```

```python
#Display X_test_cat
X_test_cat.head()
```

Code adapted from Code Reuse Session 2 – Manual Numeric Scaling (StandardScaler)

```python
# Import StandardScaler
from sklearn.preprocessing import StandardScaler

# Create scaler instance
scaler = StandardScaler()

# Fit scaler on training numeric features and transform
X_train_num = pd.DataFrame(scaler.fit_transform(X_train[numeric_cols]), columns=numeric_cols, index=X_train.index)

# Transform test numeric features using the same scaler
X_test_num = pd.DataFrame(scaler.transform(X_test[numeric_cols]), columns=numeric_cols, index=X_test.index)
```

```python
#Show X_train_num
X_train_num.describe()
```

Code adapted from Code Reuse Session 2 – Combine numeric and encoded categorical features

```python
# Concatenate numeric and one-hot categorical parts to form final training and test sets
X_train_proc = pd.concat([X_train_num, X_train_cat], axis=1)
X_test_proc  = pd.concat([X_test_num,  X_test_cat], axis=1)

# Ensure columns align exactly
X_train_proc, X_test_proc = X_train_proc.align(X_test_proc, join='outer', axis=1, fill_value=0)

# Display final shapes
X_train_proc.shape, X_test_proc.shape
```

Code adapted from Code Reuse Session 2 – Save processed datasets

```python
# Save processed datasets to CSV for later reproducibility
X_train_proc.to_csv('X_train_proc.csv', index=False)
X_test_proc.to_csv('X_test_proc.csv', index=False)
y_train.to_csv('y_train.csv', index=False)
y_test.to_csv('y_test.csv', index=False)
```

Code adapted from Code Reuse Session 2 – Train Base Classifiers

```python
# Import classifiers
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Initialize Logistic Regression (increase max_iter if needed)
lr = LogisticRegression(max_iter=2000, random_state=42)

# Fit logistic regression
lr.fit(X_train_proc, y_train)
```

```python

# Initialize Gaussian Naive Bayes
nb = GaussianNB()

# Fit GaussianNB
nb.fit(X_train_proc, y_train)
```

```python
# Initialize KNN with k=5 as baseline
knn = KNeighborsClassifier(n_neighbors=5)

# Fit KNN
knn.fit(X_train_proc, y_train)
```

Code adapted from Code Reuse Session 2 – Evaluate Classifiers (confusion matrix, classification report, AUC)

```python
# Import evaluation metrics
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve
```

```python
# Logistic Regression predictions and probabilities
y_pred_lr = lr.predict(X_test_proc)
y_proba_lr = lr.predict_proba(X_test_proc)[:,1]
```

```python
# GaussianNB predictions and probabilities
y_pred_nb = nb.predict(X_test_proc)
y_proba_nb = nb.predict_proba(X_test_proc)[:,1]
```

```python
# KNN predictions and probabilities
y_pred_knn = knn.predict(X_test_proc)
y_proba_knn = knn.predict_proba(X_test_proc)[:,1]
```

```python
# Logistic Regression: confusion matrix
print("Logistic Regression confusion matrix")
print(confusion_matrix(y_test, y_pred_lr))
print()
# GaussianNB: confusion matrix
print("GaussianNB confusion matrix ")
print(confusion_matrix(y_test, y_pred_nb))
print()
# KNN: confusion matrix
print("KNN confusion matrix")
print(confusion_matrix(y_test, y_pred_knn))
```

```python
# Logistic Regression classification report
print("Logistic Regression classification report")
print(classification_report(y_test, y_pred_lr))


# GaussianNB classification report
print("GaussianNB classification report")
print(classification_report(y_test, y_pred_nb))


# KNN classification report
print("KNN classification report")
print(classification_report(y_test, y_pred_knn))
```

```python
#Logistic Regression AUC-ROC Curve
print("AUC-ROC-Logistic Regression:", roc_auc_score(y_test, y_proba_lr))
print()
#GaussianNB  AUC-ROC Curve
print("AUC-ROC-GaussianNB:", roc_auc_score(y_test, y_proba_nb))
print()
#KNN AUC-ROC Curve
print("AUC-ROC-KNN:", roc_auc_score(y_test, y_proba_knn))
```

Code adapted from Code Reuse Session 2 – GridSearchCV for hyperparameter tuning

```python
# Import GridSearchCV and scoring utilities
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, recall_score

# Create scorer that targets the reject class recall (change pos_label if your reject label is 0 or 1)
reject_recall_scorer = make_scorer(recall_score, pos_label=0)
```

```python
# Define parameter grid for Logistic Regression
param_grid_lr = {'C':[0.01,0.1,1,10], 'class_weight':[None,'balanced']}

# Create GridSearchCV using X_train_proc and y_train
grid_lr = GridSearchCV(LogisticRegression(max_iter=2000, random_state=42),
                       param_grid=param_grid_lr,
                       cv=5,
                       scoring=reject_recall_scorer,
                       n_jobs=-1)
```

```python
# Fit GridSearch
grid_lr.fit(X_train_proc, y_train)

# Print best params and best score
print("Best LR params:", grid_lr.best_params_)
print("Best LR CV score (reject recall):", grid_lr.best_score_)
```

```python
# Define parameter grid for GaussianNB
param_grid_nb = {
    'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6]
}

# Create GridSearchCV object
grid_nb = GridSearchCV(
    GaussianNB(),
    param_grid=param_grid_nb,
    cv=5,
    scoring=reject_recall_scorer,
    n_jobs=-1
)
```

```python
# Fit GridSearch on training data
grid_nb.fit(X_train_proc, y_train)

# Print best parameters
print("Best NB parameters:", grid_nb.best_params_)
# Print best cross-validation score
print("Best NB CV score:", grid_nb.best_score_)
```

```python
# Define parameter grid for KNN
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance']
}

# Initialize GridSearchCV for KNN
grid_knn = GridSearchCV(
    estimator=KNeighborsClassifier(),
    param_grid=param_grid_knn,
    scoring=reject_recall_scorer,
    cv=5,
    n_jobs=-1
)
```

```python
# Fit GridSearchCV on training data
grid_knn.fit(X_train_proc, y_train)

# Print best parameters and CV score
print("Best KNN parameters:", grid_knn.best_params_)
print("Best KNN CV Reject Recall:", grid_knn.best_score_)
```

Code adapted from Code Reuse Session 2 – Evaluate best estimators on test set

```python
# Use best estimators from grid search if available
best_lr = grid_lr.best_estimator_

# Make predictions on the test set
y_pred_best_lr = best_lr.predict(X_test_proc)

# Get predicted probabilities for AUC (probability of positive class)
y_proba_best_lr = best_lr.predict_proba(X_test_proc)[:, 1]
```

```python
# Print confusion matrix for LR
print("Logistic Regression - Confusion Matrix")
print(confusion_matrix(y_test, y_pred_best_lr))

# Print classification report for LR
print("\nLogistic Regression - Classification Report")
print(classification_report(y_test, y_pred_best_lr))

# Print AUC-ROC for LR
print("\nLogistic Regression - AUC-ROC:", roc_auc_score(y_test, y_proba_best_lr))
```

```python
# Gaussian Naive Bayes (best from GridSearch)
best_nb = grid_nb.best_estimator_

# Make predictions on the test set
y_pred_best_nb = best_nb.predict(X_test_proc)

# Get predicted probabilities for AUC
y_proba_best_nb = best_nb.predict_proba(X_test_proc)[:, 1]
```

```python
# Print confusion matrix for NB
print("\nGaussianNB - Confusion Matrix")
print(confusion_matrix(y_test, y_pred_best_nb))

# Print classification report for NB
print("\nGaussianNB - Classification Report")
print(classification_report(y_test, y_pred_best_nb))

# Print AUC-ROC for NB
print("\nGaussianNB - AUC-ROC:", roc_auc_score(y_test, y_proba_best_nb))
```

```python
# K-Nearest Neighbours (best from GridSearch)
best_knn = grid_knn.best_estimator_

# Make predictions on the test set
y_pred_best_knn = best_knn.predict(X_test_proc)

# Get predicted probabilities for AUC
y_proba_best_knn = best_knn.predict_proba(X_test_proc)[:, 1]
```

```python
# Print confusion matrix for KNN
print("\nKNN - Confusion Matrix")
print(confusion_matrix(y_test, y_pred_best_knn))

# Print classification report for KNN
print("\nKNN - Classification Report")
print(classification_report(y_test, y_pred_best_knn))

# Print AUC-ROC for KNN
print("\nKNN - AUC-ROC:", roc_auc_score(y_test, y_proba_best_knn))
```

Code adapted from Code Reuse Session 2 – Voting Classifier Ensemble

```python
# Import VotingClassifier
from sklearn.ensemble import VotingClassifier

# Example: combine best LR and KNN (replace with your two best models)
ensemble = VotingClassifier(estimators=[('lr', best_lr), ('knn', knn)],
                            voting='soft', n_jobs=-1)

# Fit the ensemble
ensemble.fit(X_train_proc, y_train)
```

```python
# Make predictions
y_pred_ensemble = ensemble.predict(X_test_proc)

# Get probabilities
y_proba_ensemble = ensemble.predict_proba(X_test_proc)[:,1]
```

```python
# Import metrics
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score
```

```python
# Confusion matrix
print("Confusion Matrix")
print(confusion_matrix(y_test, y_pred_ensemble))

# Classification report
print(classification_report(y_test, y_pred_ensemble))

# AUC score
print("AUC:", roc_auc_score(y_test, y_proba_ensemble))
```
