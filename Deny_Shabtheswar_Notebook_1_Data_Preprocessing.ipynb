Code adapted from Code Reuse Session 1 – Correcting Logical Errors


```python
# Replace negative loan_interest_rate with median
df.loc[df['loan_interest_rate'] < 0, 'loan_interest_rate'] = df['loan_interest_rate'].median()

# Remove rows with negative max_allowed_loan
df = df[df['max_allowed_loan'] >= 0]
```

```python
#describe loan_interest_rate, max_allowed_loan
df[['loan_interest_rate','max_allowed_loan']].describe()
```

```python
#Handle Extreme Statistical Outliers

#Cap income at 99th percentile
income_cap = df['income'].quantile(0.99)
df.loc[df['income'] > income_cap, 'income'] = income_cap

#Cap loan_amount at 99th percentile
loan_cap = df['loan_amount'].quantile(0.99)
df.loc[df['loan_amount'] > loan_cap, 'loan_amount'] = loan_cap
```

```python
# verify after handle Extreme Statistical Outliers
df[['income','loan_amount']].describe()
```

```python
# verify after handle Extreme Statistical Outliers
Income_fig = px.box(df, x='income')
Income_fig.show()
```
